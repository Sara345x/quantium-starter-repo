```python
import pandas as pd
import glob
```

```python
# csv files in data folder
csv_files = glob.glob("data/*.csv")
```

```python
dfs = [] # dataframes for csv files

# Processing
for file in csv_files:
    df = pd.read_csv(file)
    
    # Filter Pink Morsels
    df = df[df['product'] == 'pink morsel'].copy()

    # Create 'Sales' column
    df['price'] = (
        df['price']
        .replace(r'[\$,]', '', regex=True)
        .astype(float)
    )
    df['Sales'] = df['quantity'] * df['price']

    df = df[['Sales', 'date', 'region']]
    df.rename(columns={'date': 'Date', 'region': 'Region'}, inplace=True)
    dfs.append(df)

# Combine files
final_df = pd.concat(dfs, ignore_index=True)
final_df.to_csv("data/formatted_output.csv", index=False)
print("Successfully processed data and saved to 'data' folder.")
```
